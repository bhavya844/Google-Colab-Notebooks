#### 2.2.1 Restaurants by City üåÜ (1 point)

For each city in california, find

1. the number of restaurants in that city,
2. mean `rating` of all restaurants in the city,
3. mean `price_range_id` per city,
4. mean `maximum_days_advance_for_reservation` per city

sort by number of restaurants.

The resulting dataframe's top 5 rows would look as follows:

| city          | restaurant_count | avg_rating | avg_price_range_id | avg_reservation_advance |
| :------------ | ---------------: | ---------: | -----------------: | ----------------------: |

#### 2.3.2 Clean up HTML text ü•ú (2 points)

Find columns containing text / strings that have html text and remove those HTML texts

ex:

```html
<p>
  Feast on delicious grub at Jerry's Famous Deli.<br />
  Its retro-style casual setting features comfortable booth seating.
</p>
```

to:

```
Feast on delicious grub at Jerry's Famous Deli. Its retro-style casual setting features comfortable booth seating.
```


````python
# ‚úèÔ∏è YOUR CODE HERE

# find the columns containing HTML tags
from bs4 import BeautifulSoup
df_new=df.copy()
cols_with_html = []
for col in df_new.columns:
    for cell in df_new[col]:
        if '<' in str(cell):
            cols_with_html.append(col)
            break

# print the columns that contain
print(cols_with_html)

# convert the columns that need to be parsed into string
for x in cols_with_html:
  df_new[x]=df_new[x].astype(str)

# remove HTML tags from the columns containing them

def htmlParser(text):
  parser=BeautifulSoup(text,'html.parser')
  return parser.get_text()

for col in cols_with_html:
  df_new[col]=df_new[col].apply(htmlParser)

# giving the output that does not contain html tags in the column
df_new.head()
````

##### 2.3.4b On my radar üó∫Ô∏è

For the following restaurant:

- brand_name `Calzone's Pizza Cucina`
- coordinates `37.799068, -122.408226`.

Answer these questions:

- How many restaurants exist within a 100 meter radius of this restaurant?
- What is the most frequent cuisine (`category`) occurence in this 100m radius across the restaurants that exist in that range?

````python
# # ‚úèÔ∏è YOUR CODE HERE

# importing geodesic from geopy.distance

from geopy.distance import geodesic

coordinates=[37.799068, -122.408226]
df_new=df.copy()

# converting latitude and longitude into a list

all_points=list(zip(df_new.latitude,df_new.longitude))

# checking how many restaurants are there within the radius of 100m

count=0
for x in all_points:
  if geodesic(coordinates,x).m <=100:
    count=count+1


print(count)

# the count 21 also consists of the restaurant itself

# creating a new column in the dataframe and adding the distances of the restaurants from the reference restaurant

df_new['distance'] = df_new.apply(lambda row: geodesic((row['latitude'], row['longitude']), (37.799068,-122.408226 )).meters, axis=1)

# there are 21 restaurants within 100 points from the given restaurant

# we filter the restaurants within a 100 meter radius
df_100m = df_new[df_new['distance'] <= 100]

# now we count the number of restaurants within the 100 meter radius
num_restaurants = len(df_100m)

# we will find the most frequent cuisine among the restaurants in the 100 meter radius
most_frequent_cuisine = df_100m['categories'].value_counts().idxmax()

#we print the most frequent cuisine

print(most_frequent_cuisine)
````
